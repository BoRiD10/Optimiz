# Mетоды антиградиента и ожига

### Импортируемые библиотеки:

```python
from sympy import *
import random
import math
import plotly.graph_objects as go
import numpy as np
```

### Функция:

```python
def Function(X, Y):
    F = 0.5 * (X ** 2 + Y ** 2) * (
                1.6 + 0.8 * cos(1.5 * X) * cos(3.14 * Y) + 0.8 * cos(math.sqrt(5) * X) * cos(3.5 * Y))
    return float(F)
```

### Метод антиградиента:

```python
def Grad (Xd, Yd, a, count):
    X, Y = symbols('x y')
    F = 0.5 * (X ** 2 + Y ** 2) * (1.6 + 0.8 * cos(1.5 * X) * cos(3.14 * Y) + 0.8 * cos(math.sqrt(5) * X) * cos(3.5 * Y))
    Dfx = diff(F, X)
    Dfy = diff(F, Y)
    Fx = Dfx.subs([(X, Xd),(Y, Yd)])
    Fy = Dfy.subs([(X, Xd),(Y, Yd)])
    if math.fabs(Fx) < 0.0005 and math.fabs(Fy) < 0.0005:
        print("Решение:\nX = " + str(Xd) + "; Y = " + str(Yd))
        print("Значение функции: \nF = " + str(F.subs([(X, Xd),(Y, Yd)])))
    else:
        Xd_new = Xd - a * Fx
        Yd_new = Yd - a * Fy
        if F.subs([(X, Xd_new),(Y, Yd_new)]) > F.subs([(X, Xd), (Y, Yd)]):
            Xd_new = Xd
            Yd_new = Yd
            a = a / 2
            count = 0
        elif round(F.subs([(X, Xd_new),(Y, Yd_new)]), 5) == round(F.subs([(X, Xd), (Y, Yd)]), 5):
            count += 1
            if count == 3:
                Xd_new = random.uniform(-2.5, 2.5)
                Yd_new = random.uniform(-2.5, 2.5)
                a = 1
                print("Выметайся отсюда!!")
        print("Промежуточное решение:\nX = " + str(Xd_new) + "; Y = " + str(Yd_new) + "; a = " + str(a))
        print("Функция: " + str(F.subs([(X, Xd_new),(Y, Yd_new)])))
        Grad(Xd_new, Yd_new, a, count)
```

### Метод ожига

```python
def Ozig(X, Y, Tmin, T):
    F = Function(X, Y)
    if T > Tmin:
        X_new = X + random.uniform(-2, 2)
        Y_new = Y + random.uniform(-2, 2)
        F_new = Function(X_new,Y_new)
        print(F_new)
        delE = F_new - F
        if delE < 0:
            F = F_new
            X = X_new
            Y = Y_new
        else :
            P = math.exp(-delE / T)
            randValue = 0.95
            if P > randValue:
                F = F_new
                X = X_new
                Y = Y_new
        T = T*0.95
        print("Промежуточное решение:\nX = " + str(X) + "; Y = " + str(Y) + "; T = " + str(T))
        print("Функция: " + str(F))
        Ozig(X, Y, Tmin, T)
    else:
        print("Решение:\nX = " + str(X) + "; Y = " + str(Y))
        print("Значение функции: \nF = " + str(F))
```

### График

```python
xs = np.arange(-2.5, 2.5, 0.1)
ys = np.arange(-2.5, 2.5, 0.1)

def XYZcreator(xs, ys):
    X, Y = np.meshgrid(xs,ys)
    Z = 0.5 * (X * X + Y * Y) * (1.6 + 0.8 * np.cos(1.5 * X) * np.cos(3.14 * Y) +
                                             0.8 * np.cos(np.sqrt(5) * X) * np.cos(3.5 * Y))
    if (xs.shape == (1,) and ys.shape == (1,)):
        return Z[0]
    else:
        return X, Y, Z


X, Y, Z = XYZcreator(xs, ys)

x_dot = np.array([0])
y_dot = np.array([0])
z_dot = XYZcreator(x_dot, y_dot)

# Блок для поверхности
surface = go.Surface(
    opacity=0.75,
    x=X,
    y=Y,
    z=Z,
    )
# Блок для точки
dot = go.Scatter3d(
    x=x_dot,
    y=y_dot,
    z=z_dot,
    mode='markers',
    marker=dict(
        size=3,
        line=dict(
            color='rgb(111, 203, 1)',
            width=0.5)
    )
)
data = [surface, dot]
layout = go.Layout(title='My func and dot', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig = go.Figure(data=data, layout=layout)

fig.show()
```
